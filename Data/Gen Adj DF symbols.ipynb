```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
symbol_adj = pd.read_csv('semart_symbol_graph.csv', header=0, index_col=0)
symbol_adj
```

```python
relevant_symbols = pd.read_csv('global_canon_names_filtered_visual.csv', index_col=0)
relevant_symbols = [x.upper() for x in relevant_symbols[relevant_symbols['visual'] == 1]['names']]
df_filtered = symbol_adj.loc[symbol_adj.index.isin(relevant_symbols)]
df_filtered = df_filtered.loc[:, symbol_adj.index.isin(relevant_symbols)]
df_filtered
```

```python
def gen_symbol_df(my_csv):
    symbol_painting = pd.DataFrame(np.zeros((my_csv.shape[0], df_filtered.shape[1])), index=my_csv['IMAGE_FILE'], columns=relevant_symbols)

    for ix, description in enumerate(my_csv['DESCRIPTION']):
        for jx, symbol in enumerate(relevant_symbols):
            if symbol.lower() in description.lower().split():
                symbol_painting.iloc[ix, jx] = 1
    
    return symbol_painting

symbol_painting_train = gen_symbol_df(pd.read_csv('/home/javierfumanal/Documents/GitHub/SemArt/semart_train.csv', delimiter='\t', encoding='Cp1252'))
symbol_painting_val = gen_symbol_df(pd.read_csv('/home/javierfumanal/Documents/GitHub/SemArt/semart_val.csv', delimiter='\t', encoding='Cp1252'))
symbol_painting_test = gen_symbol_df(pd.read_csv('/home/javierfumanal/Documents/GitHub/SemArt/semart_test.csv', delimiter='\t', encoding='Cp1252'))
```

```python
symbol_painting_train.to_csv('train_symbol_labels.csv', index=True, header=True)
symbol_painting_val.to_csv('val_symbol_labels.csv', index=True, header=True)
symbol_painting_test.to_csv('test_symbol_labels.csv', index=True, header=True)
```

```python
counts = symbol_painting_train.sum(axis=0)
average = symbol_painting_train.mean(axis=0)
```

```python
symbol_importances = pd.DataFrame(np.zeros((counts.shape[0], 2)), index=symbol_painting_train.columns, columns=['Counts', 'Average'])
symbol_importances['Counts'] = counts.astype(int)
symbol_importances['Average'] = average
symbol_importances
```
