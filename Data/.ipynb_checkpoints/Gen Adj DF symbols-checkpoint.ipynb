```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
symbol_adj = pd.read_csv('semart_symbol_graph.csv', header=0, index_col=0)
symbol_adj
```

```python
import seaborn as sns
sns.heatmap(np.log(symbol_adj+ 1))
```

```python
from sklearn import decomposition

# Scale the data
data_scaled = (symbol_adj - symbol_adj.mean()) / symbol_adj.std()

# Perform PCA
pca = decomposition.PCA(n_components=2)
pca.fit(data_scaled)
scores = pca.transform(data_scaled)

# Create a biplot
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(scores[:, 0], scores[:, 1], alpha=0.5)
for i, feature in enumerate(symbol_adj.columns):
    ax.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], head_width=0.05, head_length=0.1, fc='k', ec='k')
    ax.text(pca.components_[0, i] * 1.15, pca.components_[1, i] * 1.15, feature, color='k', ha='center', va='center')
# align(fig, ax)
plt.show()
```
